```python
import datetime as dt
import pandas as pd
import yfinance as yf
```

```python
tickers = ["AMZN","META","TSLA", "GOOG"]
start = dt.datetime.today() - dt.timedelta(3650)
end = dt.datetime.today()
cl_price = pd.DataFrame()
```

```python
for ticker in tickers:
    cl_price[ticker] = yf.download(ticker,start,end)["Adj Close"]
```

```python
cl_price.dropna(axis = 0, how = "any", inplace = True)
```

```python
daily_return = cl_price.pct_change()
```

```python
cl_price.plot()
```

```python
cl_price.plot(subplots = True, layout= (2,2))
```

```python
cl_price.plot(subplots = True, layout= (2,2), title = "Stock Price Evolution")
```

```python
cl_price.plot(subplots = True, layout= (2,2), title = "Stock Price Evolution", grid = True)
```

```python
daily_return.plot(subplots = True, layout= (2,2), title = "Stock Price Evolution", grid = True)
```

```python
(1 + daily_return).cumprod().plot()
(1+daily_return).cumsum()
```
